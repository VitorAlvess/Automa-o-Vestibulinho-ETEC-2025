```python
from selenium import webdriver
# from webdriver_manager.chrome import ChromeDriverManager para usar o chrome 
# from selenium.webdriver.chrome.service import Service

#FIrefox
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.firefox.service import Service as FirefoxService
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.common.by import By

from selenium.webdriver.common.action_chains import ActionChains
from time import sleep 
import json
import os



# Configurações do Firefox
firefox_capabilities = DesiredCapabilities.FIREFOX.copy()
firefox_capabilities['acceptInsecureCerts'] = True

options = FirefoxOptions()

firefox_profile = webdriver.FirefoxProfile()
firefox_profile.set_preference("browser.download.folderList", 2)  # Use 2 para usar o diretório personalizado
firefox_profile.set_preference("browser.download.dir", f'{os.getcwd()}\\relatorios')  # Caminho do diretório

options.set_preference("browser.download.folderList", 2)
options.set_preference("browser.download.dir", f'{os.getcwd()}\\relatorios')
options.set_preference("browser.helperApps.neverAsk.saveToDisk", "application/pdf")  # Tipos de arquivos para download automático
options.set_preference("pdfjs.disabled", True)  # Desativar visualizador interno de PDFs
options.set_preference("browser.download.useDownloadDir", True)

options.add_argument('--disable-gpu')
# options.add_argument('--headless')
options.add_argument('--window-size=1920,1080')
options.add_argument('--no-sandbox')
# options.add_argument('--start-maximized')
options.add_argument('--disable-setuid-sandbox')

# Iniciar o navegador
navegar = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()), options=options)

# Navegar até a página e fazer login
pagina = navegar
pagina.set_window_size(1366,768)
pagina.get('https://classificacao.vestibulinhoetec.com.br/')
sleep(20)  # Sleep para carregar a página

```
